```python
import numpy as np
import matplotlib.pyplot as plt

# Golden Phi constant
phi = (1 + np.sqrt(5)) / 2

# Generate phi-tuned harmonic wave
t = np.linspace(0, 2 * np.pi * phi, 1000)
wave = np.sin(t) + 0.5 * np.sin(phi * t)

# Plot resonance sieve
plt.figure(figsize=(10, 4))
plt.plot(t, wave, label='Phi-Tuned Harmonic')
plt.title('Harmonics Demo: Golden Ratio Resonance')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```
