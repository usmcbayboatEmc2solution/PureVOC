```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate mito ROS leak (entropy void)
time = np.linspace(0, 10, 100)
ros_baseline = np.exp(0.1 * time)  # Exponential leak
ros_reset = ros_baseline * np.exp(-0.2 * time)  # Cold-shock reset

# Plot mito flux
plt.figure(figsize=(10, 4))
plt.plot(time, ros_baseline, label='ROS Leak (Entropy)')
plt.plot(time, ros_reset, label='Post-Reset (Pure Center)')
plt.title('Mito Sim Demo: ROS Leak and Reset')
plt.xlabel('Time (arbitrary units)')
plt.ylabel('ROS Level')
plt.legend()
plt.grid(True)
plt.show()
```
